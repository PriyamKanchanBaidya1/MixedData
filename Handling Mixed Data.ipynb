```python
import numpy as np #for mathematical calculations
import pandas as pd #for manipulation

```

```python
df= pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df["number"].unique()
```

```python
df["number"].value_counts()
```

```python
fig=df["number"].value_counts().plot.bar()
fig.set_title("Passengers travelling in Titanic")
```

```python
df["number_numeric"]=pd.to_numeric(df["number"],errors='coerce', downcast="integer")
```

```python
df.head(20)
```

```python
df["number_categorical"]=np.where(df["number_numeric"].isnull(),df["number"],np.nan)
#this means wherever in number_numeric column there is null, take the adjacent value from number and the rest is nan
```

```python
df.head(7)
```

```python
df["Cabin"].unique()
```

```python
df["Cabin_num"]=df["Cabin"].str.extract('(\d+)')
df["Cabin_category"]=df["Cabin"].str[0]
```

```python
df.head()
```

```python
df["Cabin_category"].value_counts().plot.bar()
```

```python
df["Ticket"].unique()
```

```python
df["ticket_num"]=df["Ticket"].apply(lambda s:s.split()[-1])
df["ticket_num"]=pd.to_numeric(df["ticket_num"],errors='coerce', downcast="integer")

df["ticket_cat"]=df["Ticket"].apply(lambda s:s.split()[0])
df["ticket_cat"]=np.where(df["ticket_cat"].str.isdigit(),np.nan,df["ticket_cat"])
```

```python
df.head(20)
```

```python
df.columns
```

```python
#Drop Cabin,ticket,number
df=df.drop(['Cabin','Ticket','number'],axis=1)
```
